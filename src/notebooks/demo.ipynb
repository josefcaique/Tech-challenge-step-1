```python
import mlflow
from mlflow.tracking import MlflowClient
import mlflow.sklearn
from sklearn.linear_model import LinearRegression
from sklearn.datasets import make_regression
```

```python
mlflow.set_tracking_uri('http://mlflow:5000')
client = MlflowClient(tracking_uri="http://mlflow:5000")
```

##### Nome do Experimento

```python
experiment_name = 'demo_experimento1'
```

##### Nome do Bucket no Minio

```python
artifact_uri = "s3://mlflow-artifacts/"
```

```python
try:
    experiment_id = client.create_experiment(experiment_name, artifact_location=artifact_uri)
except FileNotFoundError:
    print('Bucket não encontrado')
    experiment_id = client.get_experiment_by_name(experiment_name).experiment_id
```

```python
mlflow.set_experiment(experiment_name)
```

```python


with mlflow.start_run():
    X,y = make_regression(n_samples =100, n_features =1, noise=0.1)
    model = LinearRegression()
    model.fit(X,y)
    
    mlflow.log_param('coeficiente', model.coef_[0])
    mlflow.log_metric("score",model.score(X,y))
    mlflow.sklearn.log_model(model,"modelo")
    
    print(f"Modelo Salvo no MLflow: {mlflow.get_artifact_uri()}")
```
